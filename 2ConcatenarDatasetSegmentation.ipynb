```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
```

```python
df_cod_ventas = pd.read_csv("sumatoria_ventas.csv")
```

```python
df_cod_ventas
```

```python
segmentacion = pd.read_csv("SegmentacionModsCatVig.csv")
```

```python
segmentacion["DIVISION"].unique()
```

```python
segmentacion_calzado = segmentacion[segmentacion["DIVISION"] == "CALZADO"]
```

```python
segmentacion_calzado
```

```python
segmentacion_calzado["SECCION"].value_counts()
```

```python
union_datasets = segmentacion_calzado.join(df_cod_ventas.set_index('ult_sust_def'), on='ULT_SUST_DEF')
```

```python
# eliminamos las columans que tienen nan en el total de ventas
union_datasets = union_datasets[pd.notna(union_datasets["Total"])]
```

```python
union_datasets.shape
```

```python
# verificamos la no repeticion de id
union_datasets["ULT_SUST_DEF"].value_counts()
```

```python
union_datasets.drop_duplicates(subset ="ULT_SUST_DEF", 
                     keep = "first", inplace = False) 
```

```python
union_datasets.to_csv("dataset_ventas_segmentacion.csv", index=False)
```

```python
df = pd.read_csv("dataset_ventas_segmentacion.csv")
```

```python
kmeans = KMeans(n_clusters=3, algorithm='elkan')
```

```python
clusters = kmeans.fit_predict(df[['Total']])
```

```python
df['Cluster'] = clusters
```

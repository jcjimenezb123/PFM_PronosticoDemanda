```python
import pandas as pd
import numpy as np
```

```python
#se cargan las ventas y existencias
df=pd.read_csv('ID_REC - 2021_11_04.csv', parse_dates=['FECHA'],dayfirst=True)
df.info(memory_usage='deep')
```

```python
df[df['VENTA']<0].count()
```

```python
df[(df['EXIST']==0)&(df['VENTA']>0)].count()
```

```python
df['VENTA']=df.apply(lambda x: np.NAN if x['EXIST'] == 0 else x['VENTA'],axis=1)
df
```

```python
#se obtiene el dia de la semana
df['dia_semana'] = df['FECHA'].dt.day_name()
#df['semana']=df['fecha'].dt.week
df['desde']=df['FECHA'].dt.to_period('W').dt.start_time
df
```

```python
#se obtiene una mustra de un modelo talla
idtalla=df[(df['ULT_SUST_DEF']==127399) & (df['TA_CVE_N']==22) & (df['TI_CVE_N']==1)]
idtalla.sort_values(by=['FECHA'],inplace=True)
```

```python
import plotly.express as px

fig = px.line(idtalla, x="FECHA", y="VENTA")
fig.add_scatter(x=idtalla['FECHA'], y=idtalla['EXIST'], mode='lines')
fig.show()
```

```python
fig = px.scatter(idtalla, x="FECHA", y="VENTA",marginal_y='histogram')
fig.show()
```
